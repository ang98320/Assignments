```python
housing_df = pd.read_excel("Median_Housing_Prices.xls")
housing_df = housing_df.filter(items=['Mon-Yr','CA', 'Los Angeles', 'San Francisco'])
```

```python
housing_df['year'] = housing_df['Mon-Yr'].dt.year
housing_df = housing_df[housing_df['year'] >= 2011]
housing_df
```
